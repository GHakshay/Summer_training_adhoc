```
import cv2
from google.colab.patches import cv2_imshow
```

```
img1=cv2.imread('1.jpg')
img2=cv2.imread('2.jpg')
```

```
img3=img1-img2# direct subtract
img4=cv2.absdiff(img1,img2) # absolute difference
```

```
cv2_imshow(img3) # direct substraction
```

```
cv2_imshow(img4) # absolute addition
```
