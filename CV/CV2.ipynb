```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```
img=cv2.imread('1.jpg')
```

```
scaleX=1
scaleY=1
```

```
scaleUp = cv2.resize(img[0:,200:300], None, fx= scaleX*3, fy= scaleY*3, interpolation= cv2.INTER_LINEAR) # specific area zoom
```

```
cv2_imshow(scaleUp)
```

```
scaleUp2 = cv2.resize(np.concatenate((img[0:,0:200],img[0:,300:]),axis=1), None, fx= scaleX*3, fy= scaleY*3, interpolation= cv2.INTER_LINEAR) # zoom in rest of the part of the image
```

```
cv2_imshow(scaleUp2)
```
